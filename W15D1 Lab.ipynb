```python
null_counts = df.isnull().sum()
null_counts
```

```python
percent_null = (null_counts / total_rows)*100
percent_null.sort_values(ascending = False)
```

```python
cols_to_drop = percent_null[percent_null > 85].index.tolist()
df1 = df.drop(columns=cols_to_drop)
df1
```
